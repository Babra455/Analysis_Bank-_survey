```python
data = pd.read_csv('cleaned_bank_survey.csv')
```

```python
def find_col(data, keywords):
    keywords = [k.lower() for k in keywords]
    for col in data.columns:
        low = col.lower()
        if all(k in low for k in keywords):
            return col
    return None

gender_col = find_col(data, ['gender'])
country_col = find_col(data, ['country'])
timestamp_col = find_col(data, ['timestamp'])
```

```python
print('gender_col =', gender_col)
print('country_col =', country_col)
```

```python
print(data[gender_col].value_counts(dropna=False).head())
print(data[country_col].value_counts(dropna=False).head())
```

```python
# Count by gender
pivot_gender = data.groupby(gender_col).size().reset_index(name='count').sort_values('count', ascending=False)

# Count by country
pivot_country = data.groupby(country_col).size().reset_index(name='count').sort_values('count', ascending=False)
```

```python
#Cross-tab: country x gender
crosstab = pd.crosstab(data[country_col], data[gender_col], dropna=False)

# Add percentage of row (within country)
crosstab_pct_row = crosstab.div(crosstab.sum(axis=1), axis=0).round(3)
```

```python
with pd.ExcelWriter('pivot_tables.xlsx', engine='xlsxwriter') as writer:
    pivot_gender.to_excel(writer, sheet_name='count_by_gender', index=False)
    pivot_country.to_excel(writer, sheet_name='count_by_country', index=False)
    crosstab.to_excel(writer, sheet_name='country_x_gender_counts')
    crosstab_pct_row.to_excel(writer, sheet_name='country_x_gender_pct_row')

print("pivot_tables.xlsx written with sheets: count_by_gender, count_by_country, country_x_gender_counts, country_x_gender_pct_row")
```
